```python
import matplotlib.pyplot as plt
import numpy as np

tests = list(range(1, 11))
detection = [820, 750, 890, 710, 840, 780, 920, 690, 860, 800]
suppression = [480, 510, 470, 490, 500, 520, 460, 510, 490, 500]
total = [1300, 1260, 1360, 1200, 1340, 1300, 1380, 1200, 1350, 1300]

x = np.arange(len(tests))
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width, detection, width, label='Detection time', color='skyblue')
bars2 = ax.bar(x, suppression, width, label='Suppression activation', color='lightgreen')
bars3 = ax.bar(x + width, total, width, label='Total response time', color='salmon')

ax.set_xlabel('Test Number')
ax.set_ylabel('Time (milliseconds)')
ax.set_title('Response Time Measurements (Fire Detection & Suppression)')
ax.set_xticks(x)
ax.set_xticklabels(tests)
ax.legend()

# Add value labels on bars
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

# Add horizontal line for average total
avg_total = np.mean(total)
ax.axhline(y=avg_total, color='red', linestyle='--', label=f'Avg Total = {avg_total:.0f} ms')
ax.legend()

plt.tight_layout()
plt.savefig('response_time_chart.png', dpi=300)
plt.show()
```
